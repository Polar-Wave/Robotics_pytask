Question 4:

```python
z = np.array([1,2,3])
x = np.array([4,5,6])

print("Addition of both arrays:",z+x)
print("Subtraction of both arrays:",z-x)
print("Multiplication of both arrays:",z*x)
print("Division of both arrays:",z/x)
print("Power of 3 of z:",z**3)
print("Power of 3 of x:",x**3)
print("Square of z:",z**2)
print("Square of x:",x**2)
print("mean of z:",np.mean(z))
print("mean of x:",np.mean(x))
```

Question 8:

```python
a = np.array([1,2,3,4])
print("original type:",a.dtype)

a_f= a.astype(float)

print(a_f)
print("\nnew type:",a_f.dtype)
```

Question 9:

```python
""" dtype is only used to check the datatype of the array and reads the datatype only but
    astype() is used to change the datatype of the array and it creates a new array with the new datatypw """
```

Question 10:

```python
mat = np.random.randint(1,100,size=(3,3))

print("Original matrix:\n", mat)

print("\nShape(rows,columns):", mat.shape)
print("Number of elements:", mat.size)
print("Number of axes:", mat.ndim)
print("Datatype:", mat.dtype)

mat_f = mat.astype(float)
print("\nNew datatype:", mat_f.dtype)
print("New matrix:\n", mat_f)
```
