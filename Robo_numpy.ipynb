```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print("Addition of both arrays:",a+b)
print("Subtraction of both arrays:",a-b)
print("Multiplication of both arrays:",a*b)
print("Division of both arrays:",a/b)
print("Power of 3 of a:",a**3)
print("Power of 3 of b:",b**3)
print("Square of a:",a**2)
print("Square of b:",b**2)
print("mean of a:",np.mean(a))
print("mean of b:",np.mean(b))
```
